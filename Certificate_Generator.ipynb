```python
import pandas as pd
from fpdf import FPDF
```

```python
# Example DataFrame
data = {
    'Name': ['John Doe', 'Jane Smith'],
    'Age': [25, 30],
    'Course': ['AI & ML', 'Data Science'],
    'Completion Date': ['2024-08-01', '2024-08-02']
}

df = pd.DataFrame(data)
```

```python
# Certificate Template Function
def create_certificate(name, age, course, completion_date, output_filename):
    pdf = FPDF('L', 'mm', 'A4')  # Landscape orientation, A4 size
    pdf.add_page()

    # Set the title of the certificate
    pdf.set_font("Arial", 'B', 24)
    pdf.cell(0, 20, 'Certificate of Completion', ln=True, align='C')

    pdf.ln(20)

    # Set the recipient's name
    pdf.set_font("Arial", 'B', 18)
    pdf.cell(0, 10, f'This certifies that', ln=True, align='C')
    pdf.ln(10)

    pdf.set_font("Arial", 'B', 24)
    pdf.cell(0, 10, name, ln=True, align='C')
    pdf.ln(10)

    pdf.set_font("Arial", '', 18)
    pdf.cell(0, 10, f'Age: {age}', ln=True, align='C')
    pdf.ln(10)

    pdf.set_font("Arial", '', 18)
    pdf.cell(0, 10, f'has successfully completed the course in {course}', ln=True, align='C')
    pdf.ln(10)

    pdf.set_font("Arial", '', 18)
    pdf.cell(0, 10, f'Completion Date: {completion_date}', ln=True, align='C')

    # Output the certificate to a file
    pdf.output(output_filename)
```

```python
# Generate certificates for each entry in the DataFrame
for index, row in df.iterrows():
    name = row['Name']
    age = row['Age']
    course = row['Course']
    completion_date = row['Completion Date']
    
    output_filename = f"certificate_{name.replace(' ', '_')}.pdf"
    create_certificate(name, age, course, completion_date, output_filename)
    print(create_certificate)
    print(f"Generated certificate for {name}")
```

```python
import fitz  # PyMuPDF

def extract_background(input_pdf, output_pdf):
    # Open the original PDF
    pdf_document = fitz.open(input_pdf)
    
    # Create a new PDF to save the background
    new_pdf = fitz.open()

    # Process each page of the original PDF
    for page_number in range(len(pdf_document)):
        # Get the page
        page = pdf_document.load_page(page_number)
        
        # Create a new page in the new PDF with the same dimensions
        new_page = new_pdf.new_page(width=page.rect.width, height=page.rect.height)
        
        # Extract the background image or graphics
        # In this case, we will extract all images as background
        for img_index, img in enumerate(page.get_images(full=True)):
            xref = img[0]
            base_image = pdf_document.extract_image(xref)
            image_bytes = base_image["image"]
            image_ext = base_image["ext"]
            # Insert the image into the new page
            new_page.insert_image(page.rect, stream=image_bytes)

        # Optionally, if your background is vector graphics, you might need to handle that separately
        # For example, you could use page.get_pixmap() to extract graphical content

    # Save the new PDF
    new_pdf.save(output_pdf)
    new_pdf.close()
    pdf_document.close()

if __name__ == "__main__":
    input_pdf = 'input.pdf'  # Replace with the path to your input PDF
    output_pdf = 'background.pdf'  # Desired output PDF with extracted background
    
    extract_background(input_pdf, output_pdf)
    print(f"Background PDF saved as {output_pdf}")
```

```python
import fitz  # PyMuPDF

def calculate_text_width(fontsize, text):
    temp_pdf = fitz.open()
    temp_page = temp_pdf.new_page()
    
    temp_page.insert_text((0, 0), text, fontsize=fontsize, render_mode=0)
    text_bbox = temp_page.get_text("dict")['blocks'][0]['lines'][0]['spans'][0]['bbox']
    text_width = text_bbox[2] - text_bbox[0]
    
    temp_pdf.close()
    
    return text_width

def truncate_text_to_fit(text, fontsize, page_width):
    words = text.split(' ')
    lines = []
    current_line = ""
    
    for word in words:
        test_line = current_line + (word + " ")
        if calculate_text_width(fontsize, test_line) <= page_width:
            current_line = test_line
        else:
            lines.append(current_line.strip())
            current_line = word + " "
    
    lines.append(current_line.strip())
    return lines

def justify_text(text, fontsize, page_width):
    lines = truncate_text_to_fit(text, fontsize, page_width)
    justified_lines = []

    for line in lines[:-1]:  # Justify all lines except the last one
        words = line.split()
        if len(words) == 1:
            justified_lines.append(line)  # No justification for single-word lines
        else:
            total_text_width = calculate_text_width(fontsize, line)
            total_space = page_width - total_text_width
            space_per_gap = total_space / (len(words) - 1)
            justified_line = (' ' * int(space_per_gap)).join(words)
            justified_lines.append(justified_line)

    justified_lines.append(lines[-1])  # Last line is not justified
    return '\n'.join(justified_lines)

def add_text_box_to_pdf(pdf_path, output_path, text_entries, positions, font_sizes, colors):
    pdf_document = fitz.open(pdf_path)
    page = pdf_document.load_page(0)
    page_width = page.rect.width - 500  # Reduce to account for padding

    for i, (text, position, fontsize, color) in enumerate(zip(text_entries, positions, font_sizes, colors)):
        if text == paragraphs_list[1]:
            text_width = calculate_text_width(fontsize, text)
            x = (page_width / 2) - text_width / 2
            y = position
        elif text == paragraphs_list[2]:
            x, y = position
            text = justify_text(text, fontsize, page_width)
        else:
            x, y = position

        page.insert_text(
            (x, y), 
            text, 
            fontsize=fontsize, 
            color=color, 
            render_mode=0
        )
    
    pdf_document.save(output_path)
    pdf_document.close()

# Define file paths and text properties
input_pdf = 'background.pdf'
output_pdf = 'background_with_text.pdf'

# Define your texts, positions, font sizes, and colors
text_entries = [paragraphs_list[0], paragraphs_list[1], paragraphs_list[2], paragraphs_list[3], paragraphs_list[4], paragraphs_list[5]]
positions = [(815, 285), 665, (100, 750), (100, 970), (80, 1400), (430, 1450)]
font_sizes = [14, 40, 25, 25, 25, 20]
colors = [(0.25, 0.25, 0.25), (0.2, 0.2, 0.2), (0, 0, 0), (0, 0, 0), (0.25, 0.25, 0.25), (0.25, 0.25, 0.25)]

# Add text boxes to the PDF
add_text_box_to_pdf(input_pdf, output_pdf, text_entries, positions, font_sizes, colors)

print(f"PDF with text boxes saved as {output_pdf}")
```
